```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error

# Generating the data with noise
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 

# Creating the model
model = Sequential()
model.add(Dense(1, input_shape=[1]))  # 1 input feature

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model.fit(x, y, epochs=1000, verbose=1)

# Making predictions
x_pred = np.linspace(4, 15, 100)
y_true = 2 * x_pred + 1
y_pred = model.predict(x_pred)
mse = mean_squared_error(y_true, y_pred)
# Plotting the data points and the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_pred, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression Example with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
print("Mean Squared Error:", mse)
```

# Traditional ML

```python
# Importing the required libraries
from sklearn.linear_model import LinearRegression

# Generating the data with noise
np.random.seed(0)
x = np.linspace(0, 10, 100).reshape(-1, 1)

y = 2 * x + 1

# Creating the linear regression model
model = LinearRegression()

# Fitting the model
model.fit(x, y)

# Making predictions
x_pred = x = np.linspace(0, 15, 100).reshape(-1, 1)
y_true = 2 * x_pred + 1
y_pred = model.predict(x_pred)
mse = mean_squared_error(y_true, y_pred)
# Plotting the data points and the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_pred, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression Example with Noise using scikit-learn')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
print("Mean Squared Error:", mse)
```

# How about add some noises?

```python
np.random.seed(0)
x = np.linspace(0, 10, 40)
noise = np.random.normal(0, 1, 40)
y = 2 * x + 1 + noise

# Creating the model
model = Sequential()
model.add(Dense(1, input_shape=[1]))  # 1 input feature

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model.fit(x, y, epochs=1000, verbose=1)

# Making predictions
x_pred = np.linspace(0, 12, 100)
noise = np.random.normal(0, 1, 100)
y_true = 2 * x_pred + 1 + noise
y_pred = model.predict(x_pred)
mse = mean_squared_error(y_true, y_pred)
# Plotting the data points and the regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_pred, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression Example with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
print("Mean Squared Error:", mse)
```
